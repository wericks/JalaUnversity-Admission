Mas... O que é programação?

Programar é dar um conjunto de instruções ao computador para que ele as execute, esse conjunto de instruções é conhecido pelo nome de `programa`, mas como passamos essas instruções para o computador? 
Bom, precisamos nos comunicar com o computador, como nos comunicamos com o computador? Vamos imaginar a comunicação entre duas pessoas para que duas pessoas possam se comunicar, elas precisam falar a mesma língua, e desta forma já podem usar voz, texto ou outras formas para realizar sua comunicação, o mesmo acontece com o computador, existem várias linguagens com as quais podemos passar as instruções para o computador, estas são conhecidas como `Linguagens de Programação` e existem vários tipos. Neste curso usaremos `Python` que é uma linguagem de programação muito popular e é usada para fazer páginas da web, jogos, incluindo aplicativos de inteligência artificial. Então, vamos escrever programas em `Python` e o computador executará nossas instruções. Vamos ver um exemplo rápido.

```python
class Dog:
  def bark(self):
    print("wof wof!")

dog = Dog()
dog.bark() 
```

Interessante! Após executar o programa, ele imprime `wof wof` no console, por enquanto, não vamos levar em consideração a estrutura do código, este é nosso primeiro exemplo digitando `Python`, pois vemos o computador executar as instruções e nos dar um resultado.

Quando os primeiros engenheiro de software e programadores surgiram, eles não entenderam nada, pois as diretrizes ou padrões não foram seguidos para escrever o código, foi aqui que o "Código Limpo" nasceu (não devemos mais mencioná-lo acima)...

Mas como é que isto é feito? ... bem usando abstração?

(exemplos de abstração) (usamos animais em vez de pessoas) (definimos um animal e suas características e comportamentos).

...uma vez que abstraímos o objeto da vida real, agora temos que representá-lo usando python.

Este é apenas o primeiro vislumbre do que veremos na próxima aula, onde falaremos sobre abstração passo a passo.

```python
class Dog:
    def __init__(self, name, dog_type):
        self.name = name
        self.dog_type = dog_type
        
    def bark(self):
        print("wof wof !")
        
```

```python
tony = Dog("tony", "golden")
tony.bark() ## Chamamos os métodos
tony.name ## Chamamos as propriedades
```
