```python
## Vamos criar a Classe Car

class Car:
    model = ""
    color = ""
    brand = ""
    
## Definimos a Classe Car com os atributos necessários
## Lembre-se que queremos apenas os atributos que usaremos no código
## e aqueles que consideramos necessários
## Vamos criar um objeto

# Instanciando um objeto
hilux = Car()
# Atribuindo valores aos atributos
hilux.model = "Hilux"
hilux.brand = "Toyota"
hilux.color = "Black"
# Imprimindo informações do objeto
print("The Car " + hilux.model + " is brand " + hilux.brand + " is of the color " + hilux.color)
```

```python
## Aqui estamos criando a Classe Car

class Car:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color
    
    def run(self):
        print("El auto " + self.model + " esta en movimiento.")
        
toyota = Car('Auris', 'Toyota', 'Cafe')
toyota.run()
```

```python
## Aqui estamos criando outro exemplo de uma classe Person.

class Person:
    
    def __init__(self, name, dni):
        self.name = name
        self.dni = dni
        
    def say_hello(self):
        print("Hola mi nombre es: " + self.name)
        
juan = Person("Juan", 1234)
juan.say_hello()

carlos = Person("Carlos", 9865)
carlos.say_hello()
print(carlos.dni)
carlos.dni = 890
print(carlos.dni)

## Para chamar atributos, precisamos apenas de ponto (.) para acessar o atributo
## Já os métodos precisamos de parênteses (), para passar parâmetros
## Não se esqueça que isso é muito importante
```

```python
class Robot:
    def __init__(self, name, color):
        self.name = name
        self.color = color
        self.age = 1
        self.batery = 8.5
        self.on = True
    
    def greet(self):
        print("My name is: ", self.name)
        print("I am ", self.age, "years old")
        print("My battery percentage is: ", self.batery)
        print("My power state is: ", self.on)


jarvis = Robot("jarvis", "red")
jarvis.greet()

ultron = Robot("ultron", "green")
ultron.greet()

robotina = Robot("robotina", "blue")
robotina.greet()
```
